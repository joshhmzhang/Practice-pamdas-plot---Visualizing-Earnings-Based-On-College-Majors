## Pandas, Histograms

Generate histograms to explore the distributions of the following columns:

```python
recent_grads['Sample_size'].hist()
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig=plt.figure(figsize=(5,20))

for i in range(8):
    ax=fig.add_subplot(8,1,i+1)
    ax.hist(recent_grads[cols[i]],bins=20)
    ax.set_title(cols[i])

        
```

## Use a box plot to explore the distributions of median salaries and unemployment rate.

```python
fig=plt.figure()
ax1=fig.add_subplot(2,1,1)
ax2=fig.add_subplot(2,1,2)
ax1.boxplot(recent_grads['Median'].values)
ax2.boxplot(recent_grads['Unemployment_rate'].values)
```
