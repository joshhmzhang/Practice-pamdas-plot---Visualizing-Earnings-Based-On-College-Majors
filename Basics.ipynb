```python
sns.kdeplot(recent_grads['Sample_size'])
```

```python
recent_grads['Sample_size'].plot.hist()
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig,axarr=plt.subplots(2,4,figsize=(20,5))

plt.subplots_adjust(hspace=.4,wspace=.2)

for i in range(4):
    recent_grads[cols[i]].plot.hist(ax=axarr[0][i],color='mediumvioletred')
    axarr[0][i].set_title(cols[i],fontsize=15)

for i in range(4,8):
    recent_grads[cols[i]].plot.hist(ax=axarr[1][i-4])
    axarr[1][i-4].set_title(cols[i],fontsize=15)
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig=plt.figure(figsize=(5,20))

for i in range(8):
    ax=fig.add_subplot(8,1,i+1)
    ax.hist(recent_grads[cols[i]],bins=20)
    ax.set_title(cols[i])

plt.subplots_adjust(hspace=.4)        
import seaborn as sns
sns.despine()
```
